# clean text function 
from duplicates, extra spaces and special characters 

```
# " ".join(foo.split())  # foo is the text 
# clean text 
import re
def cleanText(text):
  text = str(text)#enforce to string
  # transform all text to lower case 
  text = text.lower()  # transform all text to lower case first 
  #map some special characteres into text 
  text = text.replace("+", "P")
  text = text.replace("@", "AT")
  text = text.replace("#", "S")
  text = text.replace("%", "PERCENT")
  text = text.replace("&", "AND")
  text = text.replace("_", "DASH")
  text = text.replace("*", "STAR")
  text = text.replace(".", "DOT")


  #MAP NUMBERES
  text = text.replace("1", "ONE")
  text = text.replace("2", "TWO")
  text = text.replace("3", "THREE")
  text = text.replace("4", "FOUR")
  text = text.replace("5", "FIVE")
  text = text.replace("6", "SIX")
  text = text.replace("7", "SEVEN")
  text = text.replace("8", "EIGHT")
  text = text.replace("9", "NINE")
  text = text.replace("10", "TEN")







  #remove IF ANY remaining special characters and extra spaces 
  a = text
  #print("a is equal - ",a)
  for k in a.split("\n"):
    #print(re.sub(r"[^a-zA-Z0-9]+", ' ', k))
    # Or:
    string1 = " ".join(re.findall(r"[a-zA-Z0-9]+", k))
  #print("string is :",string1)
    
    
    # remove reapeatead text 
  words = string1.split() # remove repeatead text 
  #print (" ".join(sorted(set(words), key=words.index)))
  final = " ".join(sorted(set(words), key=words.index))
    
  return final
```

# similarity function

```
def similarity(tokens1, tokens2, ones_sum):
  positive_sum=0 # used to keep track of all similarity values between 1 and 0 
  tokens1 = cleanText(tokens1)
  tokens2 = cleanText(tokens2)
  tokens1 = nlp_updated(tokens1)
  tokens2 = nlp_updated(tokens2)
  tokens_len = len(tokens1) + len(tokens2)
  print("the number of tokens is ", tokens_len)
  print("from similarity function tokens1 :",tokens1)
  print("                         tokens2 :",tokens2)
  for token1 in tokens1:
      for token2 in tokens2:
          try:
            print(token1.text, token2.text, token1.similarity(token2))
            if(token1.similarity(token2) ==1):
              print("similarity is 1, add 1 to ones sum and delete token2")
              ones_sum=ones_sum + 1
              print("updated ones sum = ", ones_sum)
              to_delete = token2.text
              text1 = tokens1.text
              text2 = tokens2.text 
              tokens1, tokens2 = deleteToken(to_delete,text1,text2)
              ones_sum, positive_sum, tokens_len = similarity(tokens1, tokens2, ones_sum)
              return ones_sum, positive_sum ,tokens_len
            if(token1.similarity(token2)<0):
              print("similarity is lower than 0. ignore it ")
            if(token1.similarity(token2)>0):
              print("similarity is between 1 and 0, add it to positive sum")
              positive_sum = positive_sum + token1.similarity(token2)
              print("updated positive_sum = ", positive_sum)
          except:
              print("from expect condition error ocuured at tokens",token1,token2)
              #print(len(token1))
              #print(tokens2)
  return ones_sum, positive_sum ,tokens_len
```

#  UserWarning: [W008] Evaluating Token.similarity based on empty vectors.
https://github.com/explosion/spaCy/issues/4817

Even if a model includes word vectors in general, it doesn't mean that it necessarily has vectors available for every word that you want to query. (For instance, it's totally possible that vectors trained on data from a few years ago may not have a vector for a word like "blockchain" because it was a lot less common and maybe didn't make the frequency cutoff.)

So when you see that warning, it means that you're looking up the similarity of a token that the model doesn't have a vector for. That vector will be zero, so the similarity comparison is likely not going to be useful. To skip those words and/or find them, you can check Token.has_vector before you make the similarity comparison.
