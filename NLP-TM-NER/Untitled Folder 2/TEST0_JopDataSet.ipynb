```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('../TextFiles/csv/JopDataSet.csv')
```

```python
# to read data from the data frame at differents indexes
print(df.head())
```

```python
# to choose a specific record in the data frame(first record) 
s = df[0:1]
s=s.to_string()
print(s)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

# Create a Doc object
doc = nlp(s)

# Print each token separately
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
 # explore tokens
for token in doc:
    print(token.text, end=' | ')
```

```python
# MAKE NER 
for ent in doc.ents:
    print(ent.text, ent.start, ent.end, ent.start_char, ent.end_char, ent.label_)
```

```python
# More Specific Entities
for sent in doc.sents:
    docx = nlp(sent.text)
    if docx.ents:
        displacy.render(docx, style='ent', jupyter=True)
    else:
        print(docx.text)
```

```python
## Visualizing the entity recognizer
from spacy import displacy

displacy.render(doc, style='ent', jupyter=True)
```

```python
len([ent for ent in doc.ents ]) 
```

```python
# Viewing Specific Entities

options = {'ents': ['ORG', 'PRODUCT']}

displacy.render(doc, style='ent', jupyter=True, options=options)
```

```python
# Import the toolkit and the full Porter Stemmer library
import nltk
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer

# The Snowball Stemmer requires that you pass a language parameter
s_stemmer = SnowballStemmer(language='english')
```

```python
#Stemming

for word in re.findall(r"[^!.? ]+",s):
    print(word+' --> '+s_stemmer.stem(word))
```

```python
## Lemmatization

for token in doc:
    print(f'{token.text:{10}} {token.pos_:{10}} {token.dep_:{10}} {spacy.explain(token.dep_)}')
```
