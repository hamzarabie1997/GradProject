```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('../TextFiles/csv/home_marketing_job.csv')
```

```python
# to read data from the data frame at differents indexes

df.head()
```

```python
s = df[0:3]
s=s.to_string()
print(s)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

# Create a Doc object
doc = nlp(s)

# Print each token separately
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
 # explore tokens
for token in doc:
    print(token.text, end=' | ')
```

```python
job44 = re.findall(r"[^!.? ]+",s)
print(job44)

```

```python
# MAKE NER 

for chunk in doc.noun_chunks:
    print(chunk.text+' - '+chunk.root.text+' - '+chunk.root.dep_+' - '+chunk.root.head.text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.start, ent.end, ent.start_char, ent.end_char, ent.label_)
```

```python
## Visualizing the entity recognizer
from spacy import displacy

displacy.render(doc, style='ent', jupyter=True)
```

```python
# More Specific Entities
for sent in doc.sents:
    docx = nlp(sent.text)
    if docx.ents:
        displacy.render(docx, style='ent', jupyter=True)
    else:
        print(docx.text)
```

```python
len([ent for ent in doc.ents ]) 
```

```python
# Viewing Specific Entities

options = {'ents': ['ORG', 'PRODUCT']}
displacy.render(doc, style='ent', jupyter=True, options=options)
```

```python
#Stemming

# Import the toolkit and the full Porter Stemmer library
import nltk
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer

# The Snowball Stemmer requires that you pass a language parameter
s_stemmer = SnowballStemmer(language='english')
#Instead, we'll use another popular NLP tool called nltk, which stands for Natural Language Toolkit. 
for word in re.findall(r"[^!.? ]+",s):
    print(word+' --> '+s_stemmer.stem(word))
    
```

```python
##Function to display lemmas
def show_lemmas(text):
    for token in text:
        print(f'{token.text:{12}} {token.pos_:{6}} {token.lemma:<{22}} {token.lemma_}')
        
show_lemmas(doc)
```
